# Watershed database download
Pull the necessary HUC4 watershed boundaries for analysis.
Run this once, save the geojson file, then subsequent calls are simply reading that geojson file.

```python
import pygeohydro as pgh
import geopandas as gpd
import pandas as pd
import warnings
```

```python
wus_huc4_ids = [1503, 1605, 1606, 1701, 1702, 1703, 1705, 1706, 1707, 1708, 1709, 1710, 1711, 1712, 1801, 1802, 1803, 1804, 1805, 1806, 1807, 1808, 1809, 1810]
with warnings.catch_warnings():
    warnings.simplefilter("ignore") # Not necessary but it may spam your terminal output with multiple warnings which aren't super relevant.
    all_db = pgh.watershed.huc_wb_full(4) # This returns ALL HUC4 watershed boundaries
wus_db = all_db[all_db['huc4'].isin([str(ws) for ws in wus_huc4_ids])] # This filters just the HUC4 IDs we need. If you want to keep different HUC4 IDs, change the first line.
vars_to_keep = ['areaacres', 'areasqkm', 'states', 'huc4', 'name', 'geometry'] # You really don't need much more that these six columns, and even only the last three are necessary.
huc4_sorted = wus_db[vars_to_keep].sort_values(by='huc4', ignore_index=True) # Not super necessary to sort by numerical IDs, but it looks neater.
huc4_sorted.to_file('watersheds_WUS.geojson', driver='GeoJSON') # Save as a GeoJSON file. This is the easiest way to reuse this data.
```
